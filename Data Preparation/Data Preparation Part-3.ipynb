```python
a=[]
for i in df['District'].value_counts().index:
    a.append(i)
len(a)
```

```python
df_d=pd.read_csv(r"C:\Users\ASUS\Downloads\districts.csv")
```

```python
w=[]
for i in df['District'].value_counts().index:
    w.append(i)
```

```python
z=[]
for i in w:
    df3=df_d[(df_d['District']==i)]
    a=0
    for i in range(len(df3)-1):
        c=df3['Confirmed'][df3.index[i]]
        d=df3['Confirmed'][df3.index[i+1]]
        b=d-c
        a=a+b
    a=a//len(df3)    
    z.append(a)
    
```

```python
w=pd.DataFrame(w)
z=pd.DataFrame(z)
```

```python
w=pd.concat([w,z],axis=1)
```

```python
w.columns =['District', 'Transmission rate'] 
```

```python
df=pd.merge(df,w)
```

## Adding Deceased Rate Column

```python
w=[]
for i in df['District'].value_counts().index:
    w.append(i)
```

```python
z=[]
for i in w:
    df3=df_d[(df_d['District']==i)]
    a=0
    for i in range(len(df3)-1):
        c=df3['Deceased'][df3.index[i]]
        d=df3['Deceased'][df3.index[i+1]]
        b=d-c
        a=a+b
    a=a//len(df3)    
    z.append(a)
```

```python
z
```

```python
w=pd.DataFrame(w)
z=pd.DataFrame(z)
w=pd.concat([w,z],axis=1)
w.columns =['District', 'Deceased Rate'] 
df=pd.merge(df,w)
```

```python
df.head(5)
```

```python
for i in df['Transmission rate']:
    if i>=200:
        df['Transmission rate']=df['Transmission rate'].replace(i,'Very High')
    if i<=200 and i>50:
        df['Transmission rate']=df['Transmission rate'].replace(i,'High')
    else:
        df['Transmission rate']=df['Transmission rate'].replace(i,'Normal')
```
